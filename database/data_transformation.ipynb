```python
#Dependencies
import pymongo
import pandas as pd
import pprint
import numpy as np
```

# Extract CSV into DataFrame

```python
#Read haunted places file
loc_file = "../static/data/haunted_places.csv"
loc_df = pd.read_csv(loc_file)
loc_df.head()
```

# Clean and Transform data

```python
#Review NA columns
loc_df.count()
```

```python
#Drop NA values
loc_df.dropna(inplace=True)
```

```python
#Verify dropped values
loc_df.count()
```

```python
#Count unique cities
loc_df.groupby("city").count()
```

```python
#Count unique states
loc_df.groupby(["state", "state_abbrev"]).count()
```

```python
#Count unique country
loc_df.groupby("country").count()
```

```python
#Set counter for location ID
loc_df['loc_id'] = '' 
count = 0
for index, row in loc_df.iterrows(): 
    loc_df['loc_id'][index] = count + 1
    count +=1

#Order columns
loc_df = loc_df[['loc_id', 'location','description','city','state', 'state_abbrev', 'country', 'longitude', 'latitude', 'city_longitude', 'city_latitude']]
loc_df.head()
```

```python
#Download data to .csv for importing into database
loc_df.to_csv("../static/data/haunted_locations.csv", index=False)
```
